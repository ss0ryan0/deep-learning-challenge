## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
app_df = application_df.drop(columns=['EIN', 'NAME'])
app_df
```

```python
# Determine the number of unique values in each column
app_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_type_counts = app_df['APPLICATION_TYPE'].value_counts()
app_type_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_value = app_type_counts.median()
app_df['App_Type_Bin'] = pd.cut(app_type_counts, bins=[0, cutoff_value, float('inf')], labels=['Low', 'High'])
application_types_to_replace = app_type_counts[app_type_counts <= cutoff_value].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    app_df['APPLICATION_TYPE'] = app_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
app_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_counts = app_df['CLASSIFICATION'].value_counts()
class_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_counts[class_counts > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_value = class_counts.median()
app_df['App_Type_Bin'] = pd.cut(class_counts, bins=[0, cutoff_value, float('inf')], labels=['Low', 'High'])
classifications_to_replace = class_counts[class_counts <= cutoff_value].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    app_df['CLASSIFICATION'] = app_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
app_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
app_df_numeric = pd.get_dummies(app_df)
app_df_numeric
```

```python
# Split our preprocessed data into our features and target arrays
#refer to 20.2.05
#axis=0 is rows, axis=1 is columns
features = app_df_numeric.drop(columns=['IS_SUCCESSFUL'])
target = app_df_numeric['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
num = len(X_train_scaled[0]) #number of rows


nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=80, activation="sigmoid", input_dim=num))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=30, activation="sigmoid"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="tanh"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
nn
```

```python
#  Create a callback
import os
from tensorflow.keras.callbacks import ModelCheckpoint
os.makedirs('AlphabetSoupCharity_Optimization2/',exist_ok=True)
path = 'AlphabetSoupCharity_Optimization2.{epoch}.hdf5'
callback = ModelCheckpoint(filepath=path, save_weights_only=True, period=5)
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100, callbacks=[callback])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
tf.keras.saving.save_model(nn, 'AlphabetSoupCharity_Optimization2.h5')
```
